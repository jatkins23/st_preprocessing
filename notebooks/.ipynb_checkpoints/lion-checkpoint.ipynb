```python
from pathlib import Path
import fiona
import osmnx as ox
import pandas as pd

from st_preprocessing.locations.sources.lion import LocationsLIONLoader, LION_PATH
fiona.listlayers(LION_PATH)
```

```python
# fiona
# lion_layers = LocationsLIONLoader._load_baselayers()
import geopandas as gpd
node = gpd.read_file(LION_PATH, layer='node')
node_stname = gpd.read_file(LION_PATH, layer='node_stname')
lion = gpd.read_file(LION_PATH, layer='lion')
```

```python
import osmnx as ox
import pandas as pd

# Masks to filter LION
mask_roads = lion['FeatureTyp'].isin(['0']) # Mask only the roads
mask_road_paths = lion['FeatureTyp'].isin(['0', 'W']) # Also mask the walking paths

active_nodes = pd.concat([lion.loc[mask_road_paths,'NodeIDFrom'], lion.loc[mask_road_paths, 'NodeIDTo']]).unique().astype(int)

lion_road_edges = lion[mask_roads][['Street','FeatureTyp','NodeIDTo', 'NodeIDFrom', 'SegmentTyp', 'TrafDir', 'geometry']]
lion_road_nodes = node[node['NODEID'].isin(active_nodes)]
```

```python
lion_road_edges['NodeIDFrom'] = lion_road_edges['NodeIDFrom'].astype(int)
lion_road_edges['NodeIDTo'] = lion_road_edges['NodeIDTo'].astype(int)
lion_road_edges_indexed = (
    lion_road_edges
    .explode('geometry')
    .rename({'NodeIDFrom': 'u', 'NodeIDTo': 'v'}, axis=1)
    .set_index(['u','v'])
)

n = lion_road_edges_indexed.groupby(level=[0,1]).cumcount()
lion_road_edges_indexed = lion_road_edges_indexed.set_index(n.rename('key'), append=True)
lion_road_edges_indexed

lion_road_nodes_indexed = lion_road_nodes.rename({'NODEID':'osmid'}, axis=1).set_index('osmid')
lion_road_nodes_indexed['x'] = lion_road_nodes_indexed['geometry'].x
lion_road_nodes_indexed['y'] = lion_road_nodes_indexed['geometry'].y
```

```python
G_lion = ox.graph_from_gdfs(
    lion_road_nodes_indexed, 
    lion_road_edges_indexed
)
#ox.plot_graph(G_lion, node_size=1)
```

```python
#PROJ_CRS = str(os.getenv('PROJ_CRS'))
G_lion_consolidated = ox.consolidate_intersections(G_lion, dead_ends=False, tolerance=50)
#ox.plot_graph(G_lion_consolidated, node_size=1)
nodes, edges = ox.graph_to_gdfs(G_lion_consolidated)
nodes.explore()
```
